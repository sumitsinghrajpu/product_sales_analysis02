```python
g_data=df.groupby('ProductName')['Produc_Price'].sum()
```

```python
g_data
# plot Total Price by product
```

```python
plt.figure(figsize=(12,6))
g_data.plot(kind='bar')
plt.title("Total Price by product")
plt.xlabel("<<ProductName>>")
plt.ylabel("productPrice")
plt.xticks(rotation=45)
plt.show()
```

# LAPTOP is the product with the highest total sales among all products..

```python
max_value=g_data.max()
```

```python
max_value
```

```python
colors=['blue' if value !=max_value else 'red' for value in g_data.values]
```

```python
plt.figure(figsize=(12,6))
g_data.plot(kind='bar',color=colors)
plt.title("Total Price by product")
plt.xlabel("<<ProductName>>")
plt.ylabel("productPrice")
plt.savefig("C:/Users/ASUS/Desktop/Main_Analysis/output_images/laptop_image.png")
plt.xticks(rotation=45)
plt.show()
```
